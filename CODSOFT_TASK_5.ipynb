# Task-5 : CREDIT CARD FRAUD DETECTION
Aim : To Build a machine learning model to identify fradulent credit card transactions.

```python
# Import required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import MinMaxScaler
```

```python
# Loading the dataset
data = pd.read_csv("/content/archive (2).zip")
```

Data preprocessing and normalization



```python
# Cleaning the data
# Checking for missing values
data.isna().sum()
```

```python
# Handling missing values
data.fillna(data.mean(), inplace=True)
data.head()
```

```python
# Checking for duplicate records and remove them if necessary
print("Number of duplicate records before removal:", data.duplicated().sum())
data.drop_duplicates(inplace=True)
```

```python
# Normalize numerical data
scaler = MinMaxScaler()
data1 = scaler.fit_transform(data)
```

```python
data1
```

Handling class imbalance issues for undersampling

```python
# Handling class imbalance using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(data.drop('Class', axis=1), data['Class'])
```

Building the Model

```python
# Spliting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Training logistic regression model
log = LogisticRegression(penalty="l2",max_iter=1000,solver="lbfgs")
log.fit(X_train, y_train)
```

```python
# Make predictions
predictions = log.predict(X_test)
print(predictions)
```

```python
# Evaluating the model
accuracy = accuracy_score(y_test, predictions)
print("Accuracy:", accuracy)
print("\nClassification Report:")
print(classification_report(y_test,predictions))
```

```python
#creating confusion matrix
from sklearn.metrics import ConfusionMatrixDisplay
ConfusionMatrixDisplay.from_predictions(y_test, predictions)
```
